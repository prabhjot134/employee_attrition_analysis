```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# fetching data from the system using pandas library.

Employee_survey_data=pd.read_csv(r"D:\User Data\Desktop\nishant\foxmula\1.hr-analytics-case-study\employee_survey_data.csv")
manager_survey_data=pd.read_csv(r"D:\User Data\Desktop\nishant\foxmula\1.hr-analytics-case-study\manager_survey_data.csv")
```

# Employee survey data

```python
Employee_survey_data.head()
```

## Cheking Null Value and Treating Null Value In Employee survey data

```python
# checking the number of null fields in the data.

Employee_survey_data.info()
```

```python
# Calculating the percentage null value.

Employee_survey_data.isnull().sum()/Employee_survey_data.shape[0]
```

```python
# Calculating the median value to fill the null values.

a=Employee_survey_data.EnvironmentSatisfaction.median()
b=Employee_survey_data.JobSatisfaction.median()
c=Employee_survey_data.WorkLifeBalance.median()
r=pd.DataFrame(data={"Columns":["EnvironmentSatisfaction","JobSatisfactio","WorkLifeBalance"],"median":[a,b,c]})
print(r)
```

```python
# filling the null values with the median values.

Employee_survey_data.EnvironmentSatisfaction=Employee_survey_data.EnvironmentSatisfaction.fillna(a)
Employee_survey_data.JobSatisfaction=Employee_survey_data.JobSatisfaction.fillna(b)
Employee_survey_data.WorkLifeBalance=Employee_survey_data.WorkLifeBalance.fillna(c)

# Recheck the info

Employee_survey_data.info()
```

## Analysis Employee Suvey Data
    .Environment Satisfaction
    .Job Satisfaction
    .Work Life Balance

```python
size=Employee_survey_data.shape[0]
print("Total number of employee: ",size,"\n")
print("Average Environment Satisfaction of employee(out of 4): ",Employee_survey_data.EnvironmentSatisfaction.sum()/size)
print("Average Job Satisfaction of employee(out of 4): ",Employee_survey_data.JobSatisfaction.sum()/size)
print("Average Work Life Balance of employee(out of 4): ",Employee_survey_data.WorkLifeBalance.sum()/size)
```

```python
# Representing Employee survey data by using pie plot.

fig = plt.figure() 
ax1 = fig.add_subplot(221) 
ax2 = fig.add_subplot(222)  
ax3 = fig.add_subplot(223) 
label=["Low","Medium","high","veryHigh"]
Employee_survey_data.EnvironmentSatisfaction.value_counts().plot(kind="pie",autopct="%0.1f%%",labels=None,ax=ax1,figsize=(10,6))
ax1.set_title("Rating of Environment satisfaction by employee")
Employee_survey_data.JobSatisfaction.value_counts().plot(kind="pie",autopct="%0.1f%%",labels=None,ax=ax2,figsize=(10,6))
ax2.set_title("Rating of Job satisfaction by employee")
Employee_survey_data.WorkLifeBalance.value_counts().plot(kind="pie",autopct="%0.1f%%",labels=None,ax=ax3,figsize=(10,6))
ax3.set_title("Rating of Work Life Balance by employee")
fig.legend(labels=label,loc="lower right")
```
